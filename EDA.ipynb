```python
# prompt: grafica de barras de habitaciones

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataframe (assuming the previous code was executed)
df=pd.read_csv('https://raw.githubusercontent.com/adiacla/bigdata/refs/heads/master/vivienda_clasificacion.csv')

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
df['habitaciones'].value_counts().plot(kind='bar')
plt.title('Distribución de Habitaciones de la vivienda')
plt.xlabel('Número de Habitaciones')
plt.ylabel('Cantidad de Viviendas')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.show()
```

```python
# prompt: promedio habitaciones

# Calculate the average number of rooms
average_rooms = df['habitaciones'].mean()

print(f"El promedio de habitaciones es: {average_rooms}")
```

```python
average_rooms = df['habitaciones'].median()

print(f"La mediana de habitaciones es: {average_rooms}")
```

```python
# prompt: crea  caja de bigotes Area_parqeo

import matplotlib.pyplot as plt

# Assuming 'df' is already loaded as in the previous code
# Create the box plot
plt.figure(figsize=(8, 6))
df.boxplot(column=['Area_parqeo'])  # vert=False for horizontal boxplot
plt.title('Box Plot of Parking Area')
plt.xlabel('Area_parqeo')
plt.show()
```

```python
# prompt: borrar todos los registros que sean superior a 32 y dejarlo en el mismo df

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataframe
df=pd.read_csv('https://raw.githubusercontent.com/adiacla/bigdata/refs/heads/master/vivienda_clasificacion.csv')

# Delete rows where 'habitaciones' is greater than 32
df = df[df['Area_parqeo'] <= 31]
plt.figure(figsize=(8, 6))
df.boxplot(column=['Area_parqeo'])  # vert=False for horizontal boxplot
plt.title('Box Plot of Parking Area')
plt.xlabel('Area_parqeo')
plt.show()
```
