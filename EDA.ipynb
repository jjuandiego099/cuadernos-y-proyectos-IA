arreglos son de un solo tpo de variable(numpy) y dataframe son diferente tipos de datos(pandas)

```python
# prompt: crear un arreglo con la vaiable datos, con datos sinteticos usando numpy dodne la columna 0 sea la edad y la columna 1 sean las notas de IA que van de 0 a 5, las edaddes no pueden ser decimanles y la nota redondeala a 2 decimales

import numpy as np

# Set a seed for reproducibility
np.random.seed(42)

# Number of samples
n_samples = 20

# Generate ages (integers between 18 and 65)
edades = np.random.randint(18, 25, size=n_samples)

# Generate IA grades (floats between 0 and 5, rounded to 2 decimal places)
notas_ia = np.round(np.random.uniform(0, 5,  size=n_samples), 2)

# Create the array
datos = np.column_stack((edades, notas_ia))

datos
```

```python
# prompt: calcula el promedia de la columna 1 del arreglo de datos

# Assuming 'datos' is the NumPy array from the previous code
# Calculate the average of the first column (edades)

average_age = np.mean(datos[:, 1])

print(f"El promedio de la columna 1 (notas) es: {average_age}")
```
